# #C $\approx$ #G, #T $\approx$ #A

```python
with open('NC_002642.1.txt') as file_1:
    yaba = file_1.read().replace('\n', '')
with open('NC_001802.1.txt') as file_2:
    hiv = file_2.read().replace('\n', '')
with open('NC_045512.2.txt') as file_3:
    wuhan = file_3.read().replace('\n', '')
```

```python
letters = ['A', 'T', 'C', 'G']
yaba_occ = { x: yaba.count(x) for x in letters}
hiv_occ = { x: hiv.count(x) for x in letters}
wuhan_occ = { x: wuhan.count(x) for x in letters}
```

```python
print(yaba_occ, '\n', hiv_occ,'\n', wuhan_occ)
```

В целом соответствует утверждению.

# В некоторых геномах #C > #G в одной части и #G > #C в другой части ("GC skew")

```python
def find(genom, letter):
    return [i for i, c in enumerate(genom) if c == letter]
```

```python
yaba_c_pos = find(mn, 'C')
yaba_g_pos = find(mn, 'G')

hiv_c_pos = find(mt, 'C')
hiv_g_pos = find(mt, 'G')

wuhan_c_pos = find(nc, 'C')
wuhan_g_pos = find(nc, 'G')
```

```python
kwargs = dict(alpha=0.5, bins=100, density=True, stacked=True)

plt.hist(yaba_c_pos, **kwargs, color='g', label='C')
plt.hist(yaba_g_pos, **kwargs, color='r', label='G')

plt.gca().set(title='C amount in MN parts')
plt.legend()
```

Смотреть на остальные геномы из первой тройки не так уж интересно, слишком уж схожие графики

```python
plt.hist(hiv_c_pos, **kwargs, color='g', label='C')
plt.hist(hiv_g_pos, **kwargs, color='r', label='G')

plt.gca().set(title='C amount in AY parts')
plt.legend()
```

```python
plt.hist(wuhan_c_pos, **kwargs, color='g', label='C')
plt.hist(wuhan_g_pos, **kwargs, color='r', label='G')

plt.gca().set(title='C amount in GQ parts')
plt.legend()
```
